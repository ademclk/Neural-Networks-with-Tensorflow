```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss= 'mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
# Our relation here is Y = 3X + 1
```

```python
model.fit(xs, ys, epochs=500) # Train the neural network
```

```python
print(model.predict([10.0])) # Use the model

# As you can see it didnt predict exactly 31.0, If you increase the number of epochs then it will predict 31.
```
